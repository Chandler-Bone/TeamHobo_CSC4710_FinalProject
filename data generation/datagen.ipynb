## Generating order items

This one is a bit different, we will first need to get the order_ids from the dataframe. <br>
We will then assign 1-4 items to each order_id with the product_ids 1-10 and a quantity ranging 3-20. <br>
First we will need to get the order_ids auto generated by the database by exporting and loading back in the orders table.

```python
orders = pd.read_csv("order_exported.csv")
orders
```

```python
order_items = pd.DataFrame(columns=["order_item_id", "order_id", "product_id", "quantity"])

for id in orders["order_id"]:
    for i in range(rd.randint(1, 4)):
        product_id = rd.randint(1, 10)
        quantity = rd.randint(3, 20)
        df = pd.DataFrame([["", id, product_id, quantity]], columns=["order_item_id", "order_id", "product_id", "quantity"])
        order_items = order_items.append(df, ignore_index=True)

order_items
```

```python
order_items.to_csv('order_items.csv', index=False)
```

Now we add this to the database and all the order data is now generated.
